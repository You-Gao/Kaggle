```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Datasets

test = pd.read_csv('/kaggle/input/rsna-breast-cancer-detection/test.csv')

train = pd.read_csv('/kaggle/input/rsna-breast-cancer-detection/train.csv')
```

```python
#View the Dataset

train.head()
```

```python
# Check for Null Values

null_values = train.isna().sum()

for index in null_values.index:
    total_rows = null_values.shape[0]
    difference = null_values[index] - total_rows + 14
    print('There is a difference of {} in missing data for {}'.format(difference, index))
```

```python
# Check if there is a way to impute missing data for age

# Drop for now, come back later

train.dropna(subset=['age'], inplace = True)
```

```python
# Check How Many Patients There Are

print('There are {} amount of patients to train on'.format(train.patient_id.nunique()))

print('There are {} amount of mode pictures to train on'.format(train.groupby('patient_id')['image_id'].count().mode()[0]))
```
