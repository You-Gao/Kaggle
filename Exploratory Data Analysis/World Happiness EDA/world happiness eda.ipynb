```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
happiness_df = pd.read_csv('2019.csv')
print(happiness_df.shape)
happiness_df.head(5)
```

```python
happiness_df["Score"] = pd.cut(x = happiness_df["Score"], bins=[0, 3, 5, 7, 10], labels=["Low", "Medium", "High", "Very High"])
happiness_df.head()
```

```python
sns.barplot(x="Score", y="GDP per capita", data=happiness_df)
print(happiness_df.groupby(by="Score").mean()['GDP per capita'])
```

```python
sns.scatterplot(x="Score", y="GDP per capita", data=happiness_df)
```
