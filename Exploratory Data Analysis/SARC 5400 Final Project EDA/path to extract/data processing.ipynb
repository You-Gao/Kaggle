```python
world_indexes = pd.read_csv("World Indexes.csv")
print(world_indexes.isna().sum())
print(len(world_indexes['Country'].unique()))
world_indexes.head(5)
```

```python
world_indexes.sort_values(by='Purchasing Power Index', inplace=True, ascending=True)
world_indexes_2019 = world_indexes[world_indexes["Year"] == 2019]
```

```python
# Economic indicators
economic_indicators = ['Purchasing Power Index', 'Cost of Living Index', 'Safety Index', 'Health Care Index',]

# Humanitarian indicators
humanitarian_indicators = [ 'Property Price to Income Ratio','Traffic Commute Time Index', 'Pollution Index']
```

```python
world_indexes_2019['Climate Index'] = world_indexes_2019['Climate Index'].str.replace(',', '.').astype(float)
```

```python
# convert to percentile rank
world_indexes_2019.iloc[:, 2:-1] = world_indexes_2019.iloc[:, 2:-1].rank(pct=True)
```

```python
sns.heatmap(world_indexes_2019.iloc[:,2:-1].corr(), annot=True)
```

```python
plt.scatter(world_indexes_2019['Quality of Life Index'], world_indexes_2019['Pollution Index'])
plt.xlabel('Quality of Life Index')
plt.ylabel('Pollution Index')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))

# plot economic of egypt
sns.barplot(data=world_indexes_2019[world_indexes_2019['Country'] == 'Egypt'][economic_indicators], ax=ax[0])
ax[0].set_title('Economic/Wellness Indicators')
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=25)
ax[0].set_ylim(0, 1)

# plot humanitarian of egypt
sns.barplot(data=world_indexes_2019[world_indexes_2019['Country'] == 'Egypt'][humanitarian_indicators], ax=ax[1])
ax[1].set_title('Humanitarian Indicators')
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=25)
ax[1].set_ylim(0, 1)

plt.suptitle('Egypt Indicators')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))

# plot economic of egypt
sns.barplot(data=world_indexes_2019[world_indexes_2019['Country'] == 'Qatar'][economic_indicators], ax=ax[0])
ax[0].set_title('Economic/Wellness Indicators')
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=25)
ax[0].set_ylim(0, 1)

# plot humanitarian of egypt
sns.barplot(data=world_indexes_2019[world_indexes_2019['Country'] == 'Qatar'][humanitarian_indicators], ax=ax[1])
ax[1].set_title('Humanitarian Indicators')
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=25)
ax[1].set_ylim(0, 1)

plt.suptitle('Qatar Indicators')
```
