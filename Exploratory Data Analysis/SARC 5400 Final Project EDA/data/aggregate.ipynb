```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Ladder score 
happiness = pd.read_excel("DataForFigure2.1+with+sub+bars+2024.xlsx")
happiness["Country name"] = happiness["Country name"].str.lower()
happiness_df = happiness[["Country name", "Ladder score"]]

# EPI.new (index)
environment = pd.read_csv("epi2022results05302022.csv")
environment["country`"] = environment["country"].str.lower()
environment_df = environment[["country`", "EPI.new"]]

# S2: Refugees and IDPs, E2: Economic Inequality, P1: State Legitimacy, X1: External Intervention, P3: Human Rights
fragility = pd.read_excel("FSI-2023-DOWNLOAD.xlsx")
fragility["Country"] = fragility["Country"].str.lower()
fragility_df = fragility[["Country", "S2: Refugees and IDPs", "E2: Economic Inequality", "P1: State Legitimacy", "X1: External Intervention", "P3: Human Rights"]]

# Unnamed: 2
humandevelopment = pd.read_excel("HDR23-24_Statistical_Annex_HDI_Table.xlsx", skiprows=7)
humandevelopment["VERY HIGH HUMAN DEVELOPMENT"] = humandevelopment["VERY HIGH HUMAN DEVELOPMENT"].str.lower()
humandevelopment_df = humandevelopment[["VERY HIGH HUMAN DEVELOPMENT", "Unnamed: 2"]]

# 2021 [YR2021] | Exports
worldbank1 = pd.read_excel("P_Data_Extract_From_World_Development_Indicators (1).xlsx")
worldbank1["Country Name"] = worldbank1["Country Name"].str.lower()
worldbank1_df = worldbank1[["Country Name", "2021 [YR2021]"]]
worldbank1_df = worldbank1_df.rename(columns={"Country Name": "Country", "2021 [YR2021]": "Exports"})
worldbank1_df.dropna(inplace=True)

# 2021 [YR2021] | Imports
worldbank2 = pd.read_excel("P_Data_Extract_From_World_Development_Indicators (2).xlsx")
worldbank2["Country Name"] = worldbank2["Country Name"].str.lower()
worldbank2_df = worldbank2[["Country Name", "2021 [YR2021]"]]
worldbank2_df = worldbank2_df.rename(columns={"Country Name": "Country", "2021 [YR2021]": "Imports"})
worldbank2_df.dropna(inplace=True)

# 2022 [YR2022] | GDP
worldbank3 = pd.read_excel("P_Data_Extract_From_World_Development_Indicators (3).xlsx")
worldbank3["Country Name"] = worldbank3["Country Name"].str.lower()
worldbank3_df = worldbank3[["Country Name", "2022 [YR2022]"]]
worldbank3_df = worldbank3_df.rename(columns={"Country Name": "Country", "2022 [YR2022]": "GDP"})
worldbank3_df.dropna(inplace=True)

# 2022 [YR2022] | Consumer Price Index
worldbank4 = pd.read_excel("P_Data_Extract_From_World_Development_Indicators.xlsx")
worldbank4["Country Name"] = worldbank4["Country Name"].str.lower()
worldbank4_df = worldbank4[["Country Name", "2022 [YR2022]"]]
worldbank4_df = worldbank4_df.rename(columns={"Country Name": "Country", "2022 [YR2022]": "Consumer Price Index"})
worldbank4_df.dropna(inplace=True)

# ACCESS TO COUNTRIES
passport_power = pd.read_csv("Passport Index.csv")
passport_power = passport_power.rename(columns={"YEAR": "Year", "COUNTRY": "Country"})
passport_power["Country"] = passport_power["Country"].str.lower()
passport_power_df = passport_power[["Country", "ACCESS TO COUNTRIES", "Year"]]
passport_power_df = passport_power_df[passport_power_df["Year"] == 2022]

# Quality of Life Index,Purchasing Power Index,Safety Index,Health Care Index,Cost of Living Index,Property Price to Income Ratio,Traffic Commute Time Index,Pollution Index,Climate Index,Year
world_indexes = pd.read_csv("World Indexes.csv")
world_indexes["Country"] = world_indexes["Country"].str.lower()
world_indexes_df = world_indexes[["Country", "Quality of Life Index", "Purchasing Power Index", "Safety Index", "Health Care Index", "Cost of Living Index", "Property Price to Income Ratio", "Traffic Commute Time Index", "Pollution Index", "Climate Index", "Year"]]
world_indexes_df = world_indexes_df[world_indexes_df["Year"] == 2021]

# print the shapes of all the dfs
print(f"Happiness: {happiness_df.shape}")
print(f"Environment: {environment_df.shape}")
print(f"Fragility: {fragility_df.shape}")
print(f"Passport Power Index: {passport_power_df.shape}")
print(f"World Indexes: {world_indexes_df.shape}")
print(f"Human Development: {humandevelopment_df.shape}")
print(f"World Bank 1: {worldbank1_df.shape}")
print(f"World Bank 2: {worldbank2_df.shape}")
print(f"World Bank 3: {worldbank3_df.shape}")
print(f"World Bank 4: {worldbank4_df.shape}")

# create a list containing the union of country names with |
countries = list(set(happiness["Country name"]) | set(environment["country`"]) | set(fragility["Country"]) | set(humandevelopment["VERY HIGH HUMAN DEVELOPMENT"]) | set(worldbank1["Country Name"]) | set(worldbank2["Country Name"]) | set(worldbank3["Country Name"]) | set(worldbank4["Country Name"]) | set(passport_power["Country"]) | set(world_indexes["Country"]))
indices_df = pd.DataFrame(countries, columns=["Country"])

# merge all the dataframes
indices_df = pd.merge(indices_df, happiness_df, how="left", left_on="Country", right_on="Country name")
indices_df = pd.merge(indices_df, environment_df, how="left", left_on="Country", right_on="country`")
indices_df = pd.merge(indices_df, fragility_df, how="left", left_on="Country", right_on="Country")
indices_df = pd.merge(indices_df, humandevelopment_df, how="left", left_on="Country", right_on="VERY HIGH HUMAN DEVELOPMENT")
indices_df = pd.merge(indices_df, worldbank1_df, how="left", left_on="Country", right_on="Country")
indices_df = pd.merge(indices_df, worldbank2_df, how="left", left_on="Country", right_on="Country")
indices_df = pd.merge(indices_df, worldbank3_df, how="left", left_on="Country", right_on="Country")
indices_df = pd.merge(indices_df, worldbank4_df, how="left", left_on="Country", right_on="Country")
indices_df = pd.merge(indices_df, passport_power_df, how="left", left_on="Country", right_on="Country")
indices_df = pd.merge(indices_df, world_indexes_df, how="left", left_on="Country", right_on="Country")


# drop columns
indices_df = indices_df.drop(columns=["Country name", "country`", "VERY HIGH HUMAN DEVELOPMENT"])

# match the countries to the countries in worldbank1_df["Country"]
indices_df = indices_df[indices_df["Country"].isin(worldbank1_df["Country"])]
indices_df.reset_index(drop=True, inplace=True)
```

```python
indices_df.to_csv("indices.csv", index=False)
```

```python
country_conflicts = pd.read_csv("Country Conflicts.csv")
country_conflicts["Country"] = country_conflicts["Country"].str.lower()

fragile_states = pd.read_csv("Fragile State Index.csv")
fragile_states["Country"] = fragile_states["Country"].str.lower()
fragile_states.drop_duplicates(subset=["Country", "Year"], keep="first", inplace=True)

gender_gap = pd.read_csv("Gender Gap Index.csv")
gender_gap = gender_gap.rename(columns={"country": "Country"})
gender_gap["Country"] = gender_gap["Country"].str.lower()

passport_power = pd.read_csv("Passport Index.csv")
passport_power = passport_power.rename(columns={"YEAR": "Year", "COUNTRY": "Country"})
passport_power["Country"] = passport_power["Country"].str.lower()

world_indexes = pd.read_csv("World Indexes.csv")
world_indexes["Country"] = world_indexes["Country"].str.lower()

country_conflicts = country_conflicts.dropna()
fragile_states = fragile_states.dropna()
gender_gap = gender_gap.dropna()
passport_power = passport_power.dropna()
world_indexes = world_indexes.dropna()

print(country_conflicts.columns)
print(fragile_states.columns)
print(gender_gap.columns)
print(passport_power.columns)
print(world_indexes.columns)

print("Country Conflicts: ", country_conflicts["Year"].min(), country_conflicts["Year"].max())
print("Fragile States: ", fragile_states["Year"].min(), fragile_states["Year"].max())
print("Gender Gap: ", gender_gap["Year"].min(), gender_gap["Year"].max())
print("Passport Power: ", passport_power["Year"].min(), passport_power["Year"].max())
print("World Indexes: ", world_indexes["Year"].min(), world_indexes["Year"].max())


df1 = country_conflicts["Country"].unique()
df2 = fragile_states["Country"].unique()
df3 = gender_gap["Country"].unique()
df4 = passport_power["Country"].unique()
df5 = world_indexes["Country"].unique()
agg_countries = list(set(df1) | set(df2) | set(df3) | set(df4) | set(df5))
dfs = [country_conflicts, fragile_states, gender_gap, passport_power, world_indexes]
```

```python
aggregate_df = []
for year in range(2012, 2023):
    year_aggregate_df = pd.DataFrame(agg_countries, columns=["Country"])
    for df in dfs:
        year_df = df[df["Year"] == int(year)]
        year_df = year_df.drop(columns=["Year"])
        print(year_df.columns)
        print(year_df.shape)
        year_aggregate_df = pd.merge(year_aggregate_df, year_df, on="Country", how="left")
        print(year_aggregate_df.shape)
    aggregate_df.append(year_aggregate_df)
    year_aggregate_df["Year"] = year

aggregate_df = pd.concat(aggregate_df)
```

```python
# title case aggregate
aggregate_df["Country"] = aggregate_df["Country"].str.title()
```

```python
aggregate_df.to_csv("aggregate.csv", index=False)
```
